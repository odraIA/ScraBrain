LibriBrainSpeech

```python
from pnpl.datasets import LibriBrainSpeech

speech_example_data = LibriBrainSpeech(
    data_path="./data/",
    partition="train"
)

sample_data, label = speech_example_data[0]

# Print out some basic info about the sample
print("Sample data shape:", sample_data.shape)
print("Label shape:", label.shape)
```

```python
print("Sample data (first 10 values):", sample_data[:10])
```

```python
# create a plot of the first 5 samples and their corresponding labels
import matplotlib.pyplot as plt
fig, axs = plt.subplots(5, 1, figsize=(10, 15))
for i in range(5):
    sample_data, label = speech_example_data[i]
    axs[i].plot(sample_data)
    axs[i].set_title(f"Sample {i} - Label: {label}")
plt.tight_layout()
plt.show()
```
